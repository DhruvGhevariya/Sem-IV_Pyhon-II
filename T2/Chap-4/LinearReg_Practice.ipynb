# Task_2

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df
```

```python
df['bmi'] = np.where(df['bmi']<=18.5,'Underweight',
        np.where(df['bmi']<=25,'Normal weight',
                np.where(df['bmi']<=30,'Over-weight','Obese')))
```

```python
df
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
df = pd.get_dummies(df,drop_first=True)
df
```

```python
y = df[['charges']]
```

```python
x = df.drop('charges',axis=1)
```

```python
x
```

```python
y
```

```python
state = range(0,43)
mse_value = []
r2_value=[]

for i in state:
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
    model = LinearRegression()
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    # prediction = pd.DataFrame({
    #     'Actual':y_test.squeeze(),
    #     'Predicted':y_pred.squeeze()
    # })

    mse = mean_squared_error(y_test,y_pred)
    r2 = r2_score(y_test,y_pred)

    mse_value.append(mse)
    r2_value.append(r2)

print("MSE Values: ",mse_value)
print("R2 Values: ",r2_value)
print('Random State: ', mse_value.index(min(mse_value)),'Min MSE: ',min(mse_value))
print('Random State: ', r2_value.index(max(r2_value)),'Max MSE: ',max(r2_value))
```

```python
# After finding the best models, we predict the insurance charges for
# a new individual with specific given features (age 34, male, BMI 26, 3 children,
# smoker, southwest region).
x_test = [[34,True,False,True,False,3,True,False,False,True]]
new_y_pred = model.predict(x_test)
print(new_y_pred)
```
