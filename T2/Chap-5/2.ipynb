```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score
import seaborn as sns
```

```python
y_test = ['A','A','A','B','A','B','A','B','B','A']
y_pred = ['B','B','A','A','A','B','A','B','B','A']
# case -> A is positive

df = pd.DataFrame({
    'Actual' :y_test,
    'Predicted' : y_pred
})
df = pd.get_dummies(df,drop_first=True)
df
```

```python
conf_matrix = confusion_matrix(y_test,y_pred)
conf_matrix
```

```python
sns.heatmap(conf_matrix,annot=True,
           xticklabels=['B','A'],
           yticklabels=['B','A'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
TN = conf_matrix[0,0]
TP = conf_matrix[1,1]
FP = conf_matrix[0,1]
FN = conf_matrix[1,0]

TN,TP,FP,FN
```

```python
accuracy = (TP+TN)/np.sum(conf_matrix)
precision = TP/(TP+FP)
recall = TP/(TP+FN)

print("Accuracy : ",accuracy)
print("Precisionprecision : ",precision)
print("Recallrecall : ",recall)
```

```python
accuracy_score(df['Actual_B'],df['Predicted_B'])
```

```python
precision_score(df['Actual_B'],df['Predicted_B'])
```

```python
recall_score(df['Actual_B'],df['Predicted_B'])
```
