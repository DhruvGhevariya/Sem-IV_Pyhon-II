```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
iris = load_iris()
```

```python
x=iris.data  # .data give a x's features
y=iris.target  # .target give a x's features
```

```python
x
```

```python
y
```

```python
# test_size = 0.2
# random_state = 42
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
RFC = RandomForestClassifier(n_estimators=100,max_depth=4,max_samples=80,criterion='entropy')
```

```python
RFC.fit(x_train,y_train)
```

```python
y_pred = RFC.predict(x_test)
y_pred
```

```python
prediction = pd.DataFrame({
    'Actual':y_test.squeeze(),
    'Predicted':y_pred.squeeze()
})
prediction
```

```python
selected_tree = RFC.estimators_[0]
tree.plot_tree(selected_tree)
```

```python
accuracy_score(y_test,y_pred)
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

```python
y = df[['Outcome']]
x = df.drop('Outcome',axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=10)
```

```python
RFC = RandomForestClassifier(n_estimators=500,max_depth=4,max_samples=80,criterion='entropy')
```

```python
RFC.fit(x_train,y_train)
```

```python
y_pred = RFC.predict(x_test)
y_pred
```

```python
conf_matrix = confusion_matrix(y_test,y_pred)
conf_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
selected_tree = RFC.estimators_[0]
tree.plot_tree(selected_tree)
```
