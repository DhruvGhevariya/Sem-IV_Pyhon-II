```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df["math_pass"] = df["math score"].apply(lambda x: 1 if x >= 50 else 0)
df["reading_pass"] = df["reading score"].apply(lambda x: 1 if x >= 50 else 0)
df["writing_pass"] = df["writing score"].apply(lambda x: 1 if x >= 50 else 0)
```

```python
df.head()
```

```python
df["Overall_Pass"] = df.apply(
    lambda row: 1 if row["math_pass"] == 1 and row["reading_pass"] == 1 and row["writing_pass"] == 1 else 0,
    axis=1
)
```

```python
df.head()
```

```python
df.drop(["math score", "reading score", "writing score"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
x = df.drop("math_pass", axis=1)
y = df["math_pass"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=10
)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score
import seaborn as sns
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=10)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
