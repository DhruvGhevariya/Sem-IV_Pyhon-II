```python
data = {
    'values' :[0,40,35,45,32,39,42,100],
}
df = pd.DataFrame(data)
df
```

```python
q1 = df['values'].quantile(0.25)
q3= df['values'].quantile(0.75)

iqr = q3-q1

print(q1)
print(q3)
print(iqr)

```

```python
lower_bound = q1-1.5*iqr
upper_bound = q1+1.5*iqr

print(lower_bound)
print(upper_bound)

```

```python
outliner = df[(df['values']<lower_bound)|(df['values']>upper_bound)]
print(outliner)
```

```python
cleandf = df[(df['values']>=lower_bound)&(df['values']<=upper_bound)]
print(outliner)
```
