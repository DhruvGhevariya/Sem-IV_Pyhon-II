## Merge()


```python
employees = pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['A','B','C','D'],
    'dpt_aid':[10,20,10,30],
})

department = pd.DataFrame({
    'dpt_id':[10,20,40],
    'name':['HR','IT','Finance']
})

```

```python
pd.merge(employees,department,on='dpt_id')
```

```python
pd.merge(employees,department,on='dpt_id',how='inner')
```

```python
pd.merge(employees,department,on='dpt_id',how='outer')
```

```python
pd.merge(employees,department,on='dpt_id',how='left')
```

```python
pd.merge(employees,department,on='dpt_id',how='right')
```

```python
pd.merge(employees,department,on='dpt_id',how='right')
```

```python
pd.merge(employees,department,left_on='dpt_aid',right_on='dpt_id',how='outer',indicator = True)
```

```python
pd.merge(employees,department,left_on='dpt_aid',right_on='dpt_id',how='outer',indicator = False)
```

```python
pd.merge(employees,department,left_on='dpt_aid',right_on='dpt_id',how='outer',indicator = True,sort=True)
```

```python
pd.merge(employees,department,left_on='dpt_aid',right_on='dpt_id',how='outer',indicator = True,sort=True,copy=True)
```

```python
pd.merge(employees,department,left_on='dpt_aid',right_on='dpt_id',how='outer',indicator = True,sort=True,copy=False)
```

```python
pd.merge(employees,department,left_on='dpt_aid',right_on='dpt_id',how='outer',indicator = True,sort=True,copy=False,suffixes=['_emp','_dpt'])
```

```python
employees = pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['A','B','C','D'],
    'dpt_id':[10,20,10,30],
})

department = pd.DataFrame({
    'dpt_id':[10,20,40],
    'name':['HR','IT','Finance']
})

```

```python

pd.merge(employees,department,on='dpt_id',left_index=True)
```

```python
pd.merge(employees,department,on='dpt_id',left_index=False)
```
