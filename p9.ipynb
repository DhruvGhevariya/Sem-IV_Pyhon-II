# Networkx

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.Graph()
```

```python
G.add_node('A')
G.add_node('B')
G.add_node('C')
G.add_node('D')
```

```python
G.add_edge('A','B')
G.add_edge('B','C')
G.add_edge('C','D')
G.add_edge('D','A')
```

```python
nx.draw(G,with_labels=True)
```

```python
G1 = nx.DiGraph()
G1.add_nodes_from(['A','B','C','D','E'])
G1.add_edges_from([('A','B'),('B','C'),('C','D'),('D','E')])
nx.draw(G1,with_labels=True,node_size=1000,node_color='Green',edge_color='red',width=3)
```
