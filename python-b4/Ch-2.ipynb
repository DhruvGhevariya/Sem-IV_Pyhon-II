## HeatMap


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=np.random.randint(low=1,high=100,size=(10,10))
print(data)
# sns.heatmap(data)
hm=sns.heatmap(data,cmap="tab20")
plt.show()
```

```python
hm=sns.heatmap(data,annot=True)
plt.show()
```

```python
hm=sns.heatmap(data,annot=True,cmap="cool",linewidths=2,linecolor='k',cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

```python
import pandas as pd
df=pd.DataFrame(np.random.random((5,5)),
               columns=['a','b','c','d','e'])
print(df)
sns.heatmap(df)
```

```python
plt.figure(figsize=(15,5))
data=np.random.randint(low=1,high=100,size=(3,10))
print(data)
sns.heatmap(data,cmap="tab20")
plt.show()
```
